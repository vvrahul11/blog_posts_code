In this notebook, we will load several time series datasets and visualize each one of them using simple python functions.

# Import libraries

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from bokeh.models import ColumnDataSource
from bokeh.plotting import figure, show, output_file
from bokeh.io import output_notebook
```

```python
sns.set_style(style="whitegrid")
from matplotlib import rcParams

%matplotlib inline
plt.style.use('fivethirtyeight')
rcParams['axes.labelsize'] = 14
rcParams['xtick.labelsize'] = 18
rcParams['ytick.labelsize'] = 18
#rcParams['text.color'] = 'K'
rcParams['figure.figsize'] = 16,5

font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 22}

plt.rc('font', **font)
```

# Functions

```python
def load_time_data(path, files):
    """
    Load and plot time series data
    """
    for file in files:
        print(f"file name is:", file)
        #import pdb; pdb.set_trace()
        df = pd.read_csv(os.path.join(path, file), header=None)
        df.columns = [str(file)]
        df.plot(figsize = (25, 5))

        
def bokeh_plot(path, files):    
    """
    Load and plot time series data using bokeh
    """
    for file in files:
        print(f"file name is:", file)
        df = pd.read_csv(os.path.join(path, file), header=None)
        df.columns = [str(file)]
        col = df.columns[0]
        
        output_notebook()
        p = figure(x_axis_type="datetime", plot_width=900, plot_height=350)
        p.line(df.index, df[col], color = 'cornflowerblue')
        show(p)     
```

# Load and display data

```python
files = ['ecg-heartbeat-av.txt',
             'italianpowerdemand.csv',
             'mfcc_queen.csv', 
             'mfcc_vanilla_ice.csv', 
             'PigInternalBleedingDatasetArtPressureFluidFilled_100_7501.txt',
             'sudden-cardiac-death.txt',
             'walk-jog-run.txt']

load_time_data("data", files)       
```

```python
bokeh_plot("data", files)   
```
